```python
# Task 2.1: Email Validation Checker

def check_email(email):
    # Check if input has exactly one @
    if email.count("@") != 1:
        return "Invalid email"

    # Split the email into two parts, one before @ and one after @
    parts = email.split("@")
    user = parts[0]
    part2 = parts[1]

    # Check if there is at least one . after the @
    if "." not in part2:
        return "Invalid email"

    print("Username:", user)

    # Extract Domain:
    domain_parts = part2.split(".")
    
    # The domain name:
    domain_name = domain_parts[0]
    print("Domain:", domain_name)

    # Check for Domain Ending:
    if email.endswith(".com"):
        return "Commercial Domain"
    elif email.endswith(".edu"):
        return "Educational Domain"
    else:
        return "Other Domain"

# Testing:
my_email = "Amit_ml@gmail.edu"
result = check_email(my_email)
print("Result:", result)
```

```python

def decode_message(encoded_message):
    # Cleaning by going through all characters
    clean_message = ""
    for char in encoded_message:
        if char.isalpha() or char == " ":
            clean_message = clean_message + char
            
    # clean_message now looks like "mocleW EPGTQ"
            
    # 2. Separation Using Slicing
    
    space_index = clean_message.find(" ")
    
    # Slice the first word (0 to space)
    word1 = clean_message[0 : space_index]
    
    # Slice the second word (space + 1 to end)
    word2 = clean_message[space_index + 1 : ]

    # 3. Reverse of word 1
    reversed_word1 = word1[::-1]

    # 4. Decode word 2
    decoded_word2 = ""
    for letter in word2:
        if letter == "I":
            decoded_word2 = decoded_word2 + "E" 
        elif letter == "O":
            decoded_word2 = decoded_word2 + "U"
        elif letter == "E":
            decoded_word2 = decoded_word2 + "A"
        elif letter == "U":
            decoded_word2 = decoded_word2 + "O"
        else:
            decoded_word2 = decoded_word2 + letter

    return reversed_word1 + " " + decoded_word2

# Code check:

# TASK 2: Encoded message "###!!@mocleW EPGTQ!!!6789"
print("\nOutput of Task2")
print(decode_message("###!!@mocleW EPGTQ!!!6789"))

# TASK 3: Encoded message "&&&**$gnirts PLIO!!@1234"
print("\nOutput of Task3")
print(decode_message("&&&**$gnirts PLIO!!@1234"))

# TASK 4: Encoded message "##$$$@!yalpstcejorp EPUVT****9887"
print("\nOutput of Task4")
print(decode_message("##$$$@!yalpstcejorp EPUVT****9887"))
```
