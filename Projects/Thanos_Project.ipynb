```python
import os
import random
import time

def thanos_project(num_files_to_create):
    
    # 1. Defining the path to the desktop
    user_home = os.path.expanduser("~")
    desktop = os.path.join(user_home, "Desktop")
    folder_path = os.path.join(desktop, "Universe")
    
    print(f"--- TARGET LOCATED: {folder_path} ---")

    # 2. Creating the "Universe"
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
    
    # 3. Populating the Universe
    print(f"Creating {num_files_to_create} citizens...")
    for i in range(num_files_to_create):
        file_path = os.path.join(folder_path, f"citizen_{i}.txt")
        with open(file_path, 'w') as f:
            f.write("Perfectly balanced, as all things should be.")
    
    files = os.listdir(folder_path)
    print(f"Initial Population: {len(files)} citizens.")
    
    
    # 4. THE SNAP (Randomly delete half)
    print("*SNAP*")
    random.shuffle(files)
    
    # Calculating half the population
    victims = files[:len(files) // 2]
    
    for victim in victims:
        os.remove(os.path.join(folder_path, victim))
        print(f" dusted: {victim}")
    
    # 5. Checking the result, unlike Thanos who never checked back..
    remaining = os.listdir(folder_path)
    print(f"------------------------------------------------")
    print(f"{len(remaining)} citizens remain. Balance restored.")

# To determine how much files the universe folder contains:
if __name__ == "__main__":
    thanos_project(10)
```
