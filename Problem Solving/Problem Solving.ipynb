## CodeWars Problem Solving Series:
### Problem 1: Reversed Words

```python
def reverse_words(s):
    return " ".join(s.split()[::-1])

# --- Example Execution ---
if __name__ == "__main__":
    test_string = "The greatest victory is that which requires no battle"
    print(f"Original: '{test_string}'")
    print(f"Reversed: '{reverse_words(test_string)}'")

# METHOD 2-------------------------------------------------------------------------------------------------------

def reverse_words_alt(s):
    return " ".join(reversed(s.split()))

# --- Example Execution ---
if __name__ == "__main__":
    test_string = "Hello world from Python"
    print(f"Original: '{test_string}'")
    print(f"Reversed: '{reverse_words_alt(test_string)}'")
```

### Problem 2: Are You Playing Banjo?

```python
def are_you_playing_banjo(name):
    if name[0].lower() == 'r':
        return name + " plays banjo"
    else:
        return name + " does not play banjo"

# --- Example Execution ---
if __name__ == "__main__":
    names = ["Martin", "Ringo", "roland", "Diana"]
    for name in names:
        print(are_you_playing_banjo(name))

# METHOD 2-------------------------------------------------------------------------------------------------------

def are_you_playing_banjo_alt(name):
    if name.lower().startswith('r'):
        return f"{name} plays banjo"
    return f"{name} does not play banjo"

# --- Example Execution ---
if __name__ == "__main__":
    for name in ["Rick", "morty", "Rachel"]:
        print(are_you_playing_banjo_alt(name))
```

### Problem 3: Two Sum

```python
class Solution:
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        seen = {}
        for i, num in enumerate(nums):
            diff = target - num
            if diff in seen:
                return [seen[diff], i]
            seen[num] = i

# --- Example Execution ---
if __name__ == "__main__":
    sol = Solution()
    nums = [2, 7, 11, 15]
    target = 9
    print(f"Array: {nums}, Target: {target}")
    print(f"Indices: {sol.twoSum(nums, target)}")

# METHOD 2-------------------------------------------------------------------------------------------------------

class Solution:
    def twoSumAlt(self, nums: list[int], target: int) -> list[int]:
        for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                if nums[i] + nums[j] == target:
                    return [i, j]
        return []

# --- Example Execution ---
if __name__ == "__main__":
    sol = Solution()
    print(f"Indices: {sol.twoSumAlt([3, 2, 4], 6)}")
```

### Problem 4: Adding Two Numbers

```python
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(0)
        curr = dummy
        carry = 0
        
        while l1 or l2 or carry:
            val1 = l1.val if l1 else 0
            val2 = l2.val if l2 else 0
            
            total = val1 + val2 + carry
            carry = total // 10
            curr.next = ListNode(total % 10)
            
            curr = curr.next
            if l1: l1 = l1.next
            if l2: l2 = l2.next
            
        return dummy.next

# --- Example Execution ---
# Helper functions to test linked lists locally
def build_linked_list(arr):
    dummy = ListNode()
    curr = dummy
    for val in arr:
        curr.next = ListNode(val)
        curr = curr.next
    return dummy.next

def print_linked_list(node):
    res = []
    while node:
        res.append(str(node.val))
        node = node.next
    return " -> ".join(res)

if __name__ == "__main__":
    sol = Solution()
    # Representing 342 + 465 = 807
    l1 = build_linked_list([2, 4, 3]) 
    l2 = build_linked_list([5, 6, 4])
    
    result = sol.addTwoNumbers(l1, l2)
    print(f"List 1: {print_linked_list(l1)}")
    print(f"List 2: {print_linked_list(l2)}")
    print(f"Result: {print_linked_list(result)}")

# METHOD 2--------------------------------------------------------------------------------------------------------------------------

from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbersAlt(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        def get_num(node):
            num_str = ""
            while node:
                num_str += str(node.val)
                node = node.next
            return int(num_str[::-1]) # Reverse string and convert to int
        
        total = get_num(l1) + get_num(l2)
        
        # Convert total back to reversed linked list
        dummy = ListNode(0)
        curr = dummy
        for char in reversed(str(total)):
            curr.next = ListNode(int(char))
            curr = curr.next
            
        return dummy.next

# --- Example Execution ---
def build_linked_list(arr):
    dummy = ListNode()
    curr = dummy
    for val in arr:
        curr.next = ListNode(val)
        curr = curr.next
    return dummy.next

def print_linked_list(node):
    res = []
    while node:
        res.append(str(node.val))
        node = node.next
    return " -> ".join(res)

if __name__ == "__main__":
    sol = Solution()
    l1 = build_linked_list([2, 4, 3]) 
    l2 = build_linked_list([5, 6, 4])
    result = sol.addTwoNumbersAlt(l1, l2)
    print(f"Result: {print_linked_list(result)}")
```

### Problem 5: Converting Roman Digits to Integers

```python
class Solution:
    def romanToInt(self, s: str) -> int:
        roman = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
        total = 0
        
        for i in range(len(s)):
            if i + 1 < len(s) and roman[s[i]] < roman[s[i+1]]:
                total -= roman[s[i]]
            else:
                total += roman[s[i]]
                
        return total

# --- Example Execution ---
if __name__ == "__main__":
    sol = Solution()
    test_cases = ["III", "LVIII", "MCMXCIV"]
    for roman in test_cases:
        print(f"Roman: {roman} -> Integer: {sol.romanToInt(roman)}")

# METHOD 2-------------------------------------------------------------------------------------------------------

class Solution:
    def romanToIntAlt(self, s: str) -> int:
        roman = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
        total = 0
        prev_value = 0
        
        for char in reversed(s):
            current_value = roman[char]
            if current_value < prev_value:
                total -= current_value
            else:
                total += current_value
            prev_value = current_value
                
        return total

# --- Example Execution ---
if __name__ == "__main__":
    sol = Solution()
    print(f"MCMXCIV -> {sol.romanToIntAlt('MCMXCIV')}")
```

### Problem 6: Searching Integer Position

```python
class Solution:
    def searchInsert(self, nums: list[int], target: int) -> int:
        left, right = 0, len(nums) - 1
        
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1
                
        return left

# --- Example Execution ---
if __name__ == "__main__":
    sol = Solution()
    nums = [1, 3, 5, 6]
    targets = [5, 2, 7]
    
    for target in targets:
        index = sol.searchInsert(nums, target)
        print(f"Array: {nums}, Target: {target} -> Insert at index: {index}")

# METHOD 2-------------------------------------------------------------------------------------------------------

import bisect

class Solution:
    def searchInsertAlt(self, nums: list[int], target: int) -> int:
        return bisect.bisect_left(nums, target)

# --- Example Execution ---
if __name__ == "__main__":
    sol = Solution()
    print(f"Insert 5 into [1, 3, 5, 6] at index: {sol.searchInsertAlt([1, 3, 5, 6], 5)}")
    print(f"Insert 2 into [1, 3, 5, 6] at index: {sol.searchInsertAlt([1, 3, 5, 6], 2)}")
```

### Problem 7: Basic Calculator

```python
class Solution:
    def calculate(self, s: str) -> int:
        if not s:
            return 0
            
        stack = []
        num = 0
        sign = '+'
        
        for i, char in enumerate(s):
            if char.isdigit():
                num = num * 10 + int(char)
                
            if char in "+-*/" or i == len(s) - 1:
                if sign == '+':
                    stack.append(num)
                elif sign == '-':
                    stack.append(-num)
                elif sign == '*':
                    stack.append(stack.pop() * num)
                elif sign == '/':
                    stack.append(int(stack.pop() / num)) 
                
                sign = char
                num = 0
                
        return sum(stack)

# --- Example Execution ---
if __name__ == "__main__":
    sol = Solution()
    expressions = ["3+2*2", " 3/2 ", " 3+5 / 2 "]
    
    for expr in expressions:
        print(f"Expression: '{expr}' -> Result: {sol.calculate(expr)}")

# METHOD 2-------------------------------------------------------------------------------------------------------

class Solution:
    def calculateAlt(self, s: str) -> int:
        if not s: return 0
        
        result, tail, num = 0, 0, 0
        sign = '+'
        
        for i, char in enumerate(s):
            if char.isdigit():
                num = num * 10 + int(char)
                
            if char in "+-*/" or i == len(s) - 1:
                if sign == '+':
                    result += num
                    tail = num
                elif sign == '-':
                    result -= num
                    tail = -num
                elif sign == '*':
                    result = result - tail + (tail * num)
                    tail = tail * num
                elif sign == '/':
                    # Python's integer division truncates toward negative infinity
                    # So we use int() division to truncate toward zero
                    new_tail = int(tail / num)
                    result = result - tail + new_tail
                    tail = new_tail
                    
                sign = char
                num = 0
                
        return result

# --- Example Execution ---
if __name__ == "__main__":
    sol = Solution()
    print(f"Expression '3+2*2' -> Result: {sol.calculateAlt('3+2*2')}")
    print(f"Expression '14-3/2' -> Result: {sol.calculateAlt('14-3/2')}")
```
